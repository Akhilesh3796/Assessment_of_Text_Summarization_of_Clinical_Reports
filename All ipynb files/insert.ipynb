```python
# creating the join with multiple primary and secondary_id into result.csv

import csv
import mysql.connector

# Establish a MySQL connection
connection = mysql.connector.connect(
    host='localhost',
    user='root',
    password='root',
    database = 'clinical_data'
)


# Replace 'your_database' with the name of the database you created

# Create a cursor object
cursor = connection.cursor()

# Execute the SQL query
query = """
    SELECT *
    FROM clinical_comparisons AS t
    INNER JOIN all_clinical_files AS o ON t.Primary_id = o.Clinical_Trial_ID
    INNER JOIN all_clinical_files AS d ON t.secondary_id = d.Clinical_Trial_ID
"""
cursor.execute(query)

# Fetch all the rows from the result
rows = cursor.fetchall()

# Define the CSV file path
csv_file_path = 'C:/Users/akhil/OneDrive/Desktop/result.csv'

# Write the rows to the CSV file
with open(csv_file_path, 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow([desc[0] for desc in cursor.description])  # Write the header
    writer.writerows(rows)  # Write the rows


# Close the cursor and connection
cursor.close()
connection.close()
```

```python
#Appending all Single type files with the result.csv

import csv
import mysql.connector

# Establish a MySQL connection
connection = mysql.connector.connect(
    host='localhost',
    user='root',
    password='root',
    database = 'clinical_data'
)


# Create a cursor object
cursor = connection.cursor()

# Execute the SQL query
query = """
    SELECT *
    FROM clinical_comparisons AS t
    INNER JOIN all_clinical_files AS o ON t.Primary_id = o.Clinical_Trial_ID
    WHERE t.com_type = 'Single'
"""
cursor.execute(query)

# Fetch all the rows from the result
rows = cursor.fetchall()

# Define the CSV file path
csv_file_path = 'C:/Users/akhil/OneDrive/Desktop/result.csv'

# Write the rows to the CSV file with UTF-8 encoding
with open(csv_file_path, 'a', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow([desc[0] for desc in cursor.description])  # Write the header
    writer.writerows(rows)  # Write the rows

# Close the cursor and connection
cursor.close()
connection.close()
```
