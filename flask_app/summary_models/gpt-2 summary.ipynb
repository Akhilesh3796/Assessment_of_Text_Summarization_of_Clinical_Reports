```python
import pandas as pd
from summarizer import TransformerSummarizer  # Make sure you have the 'summarizer' library installed
from rouge_score import rouge_scorer  # For calculating ROUGE scores
def calculate_rouge(reference, hypothesis):
    scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
    scores = scorer.score(reference, hypothesis)
    return scores
def generate_summary_gpt2(text):
    GPT2_model = TransformerSummarizer(transformer_type="GPT2", transformer_model_key="gpt2-medium")
    
    summary = GPT2_model(text, min_length=200)
    return summary
```

```python
def process_csv(input_csv):
    data = pd.read_csv(input_csv)

    total_rouge_scores = {'rouge1': 0, 'rouge2': 0, 'rougeL': 0}
    num_rows = len(data)

    for index, row in data.iterrows():
        text = str(row['merged_text'])  # Convert to string if needed
        summary = generate_summary_gpt2(text)
        
        # Calculate ROUGE scores
        rouge_scores = calculate_rouge(text, summary)
        
        for metric in total_rouge_scores.keys():
            total_rouge_scores[metric] += rouge_scores[metric].fmeasure
    
    # Calculate average ROUGE scores
    avg_rouge_scores = {metric: total_score / num_rows for metric, total_score in total_rouge_scores.items()}
    
    return avg_rouge_scores


# Call the function with your CSV file
input_csv = 'd:/Thesis/Processed Data/test_data/11_data.csv'
avg_rouge_scores = process_csv(input_csv)
print("Average ROUGE Scores:", avg_rouge_scores)
```

```python
import pandas as pd
import torch
from summarizer import TransformerSummarizer
from transformers import BertTokenizer
from rouge_score import rouge_scorer

# Function to calculate Rouge score
def calculate_rouge(reference, summary):
    scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
    scores = scorer.score(reference, summary)
    return scores

# Function to generate summary using Transformer-based model
def generate_summary_transformer(text):
    transformer_model = TransformerSummarizer(transformer_type="GPT2", transformer_model_key="gpt2-medium")
     # Move the model to GPU
    
    summary = transformer_model(text, min_length=200)  # Use generate_summary method
    
    rouge_scores = calculate_rouge(text, summary[0])
    return summary[0]


# Load CSV file
input_csv = "d:/Thesis/Processed Data/test_data/11_data.csv"  # Update with your file path
output_csv = "d:/Thesis/Processed Data/test_data/gpt_final_summary.csv"  # Update with your file path

# Read CSV and summarize merged_text column
df = pd.read_csv(input_csv)
summaries = []
rouge_scores = []

for idx, row in df.iterrows():
    summary, rouge_score = generate_summary_transformer(row['merged_text'])
    summaries.append(summary)
    rouge_scores.append(rouge_score)

# Calculate average Rouge scores for each metric
avg_rouge_scores = {
    'rouge1': sum(score['rouge1'].fmeasure for score in rouge_scores) / len(rouge_scores),
    'rouge2': sum(score['rouge2'].fmeasure for score in rouge_scores) / len(rouge_scores),
    'rougeL': sum(score['rougeL'].fmeasure for score in rouge_scores) / len(rouge_scores)
}

# Add summaries and scores to the DataFrame
df['summary'] = summaries
df['rouge_scores'] = rouge_scores

# Write DataFrame back to CSV
df.to_csv(output_csv, index=False)

# Print average Rouge scores
print("Average Rouge Scores:")
for metric, value in avg_rouge_scores.items():
    print(f"{metric}: {value:.4f}")
```
